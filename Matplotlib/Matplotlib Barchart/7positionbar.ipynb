```python
'''
@Author: Rashmi
@Date: 2021-10-7 18:40
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 18:50
@Title :Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''
import matplotlib.pyplot as plt
import numpy as np

pgm_lang = ['Java','Python','PHP','JavaScript','C#','C++']
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
xpos = np.arange(len(pgm_lang)) #xpos is array 
plt.xticks(xpos,pgm_lang)
plt.xlabel("programming langs")
plt.ylabel("Popularity")
plt.title("bar chart of the popularity of programming Languages specific positins of bars")
y_pos = [1,3,6,8,10,11]  #spacing 2,4,7,9,11,12(on these indexes )
plt.xticks(y_pos,pgm_lang)
plt.bar(y_pos,Popularity,color =['cyan','yellow','green','orange','red','blue'])
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show() 
```
