```python
'''
@Author: Rashmi
@Date: 2021-10-7 15:35
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 15:59
@Title :Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).
Sample data:Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''
import matplotlib.pyplot as plt
  
# function to add value labels
def addlabels(pgm_lang ,Popularity):
    for i in range(len(pgm_lang)):
        plt.text(i, Popularity[i],float(Popularity[i]), ha = 'center')
  
if __name__ == '__main__':
    pgm_lang = ['Java','Python','PHP','JavaScript','C#','C++']
    Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
    plt.bar(pgm_lang , Popularity,color = ['cyan','yellow','green','orange','red','blue']) 
    # calling the function to add value labels
    addlabels(pgm_lang,Popularity)
    plt.title("bar chart of the popularity of programming Languages")
    plt.xlabel("programming langs")
    plt.ylabel("Popularity")
    plt.show()

```
