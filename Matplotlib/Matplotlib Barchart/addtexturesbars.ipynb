```python
'''
@Author: Rashmi
@Date: 2021-10-7 21:5
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 21:34
@Title :Write a Python program to add textures (black and white) to bars and wedges. Note: Use
bottom to stack the womens bars on top of the mens bars.
'''
import matplotlib.pyplot as plt
fig = plt.figure()
patterns = [ "x","o","O","|","+", "-","\\", ".", "*", "/"]
graph_plot = fig.add_subplot(111)
for element in range(len(patterns)):
    graph_plot.bar(element, 3, color='white', edgecolor='black', hatch=patterns[element])
plt.show()
```
