```python
'''
@Author: Rashmi
@Date: 2021-10-7 20:03
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 20:15
@Title :Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645
'''
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

menMeans = (54.74, 42.35, 67.37, 58.24, 30.25)
menStdev = (4, 3, 4, 1, 5)
index = np.arange(5)    
width = 0.35   
plt.bar(index, menMeans, width, yerr=menStdev, color='yellow')
plt.ylabel('Scores')
plt.xlabel('Velocity')
plt.title('Scores by Velocity showing mean std deviation as error bar')
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='blue')
plt.show()
```
