```python
'''
@Author: Rashmi
@Date: 2021-10-7 23:33
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 23:48
@Title :Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.
'''
import matplotlib.pyplot as plt
import numpy as np 
x_value = np.random.randn(80) 
y_value = np.random.randn(80)
plt.scatter(x_value,y_value,color = 'green',marker='+')
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.show()
```
