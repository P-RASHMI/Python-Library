```python
'''
@Author: Rashmi
@Date: 2021-10-8 7:00
@Last Modified by: Rashmi
@Last Modified time: 2021-10-8 7:11
@Title :Write a Python program to draw a scatter plot to find sea level rise in past 100 years.
'''
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('dataseries.csv')
time = data['Time']
gmsl = data['GMSL']
plt.scatter(time,gmsl,color = 'red',marker = '+')
plt.xlabel('year')
plt.ylabel('sea level')
plt.title('Rise in Sealevel')
plt.show()
```
