```python
'''
@Author: Rashmi
@Date: 2021-10-7 22:44
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 22:56
@Title :Write a Python programming to create a pie chart with a title of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
'''
import matplotlib.pyplot as plt

pmg_langs = 'Java','Python','PHP','JavaScript','C#','C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['red', 'gold', 'green', 'blue', 'lightcoral', 'lightskyblue']
plt.pie(popularity,labels = pmg_langs,shadow=True,autopct='%1.1f%%',explode=[0,0.1,0,0,0,0],colors = colors)
plt.title("Popularity of Programming Languages",bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```
