```python
'''
@Author: Rashmi
@Date: 2021-10-6 23:05
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 23:45
@Title :Write a Python program to create multiple types of charts(line,scattered plot,bar chart)
'''
import matplotlib.pyplot as plt 
#line graph
x=[1,2,3,4,5,7]
y=[2,1,6,4,8,5]

plt.plot(x,y,color = 'blue',marker = '*') 
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("line graph")
plt.show()

```

```python
'''To print scattered plot'''
import matplotlib.pyplot as plt 

x=[1,2,3,4,5,7]
y=[2,1,6,4,8,5]

plt.scatter(x,y,color = 'cyan',marker = 's') 
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("scattered plot")
plt.show()

```

```python
''' To print Bar chart '''
import matplotlib.pyplot as plt
months = ['jan','feb','march','april']
A_revenue = [15,12,18,20]
B_revenue = [21,25,44,31]
plt.xlabel("Months")
plt.ylabel("revenue")
plt.title("Monthly revenues of Company A and B,(Bar chart)")
plt.bar([0.25,1.25,2.25,3.25],A_revenue,label = "COM_A",width =0.4,color = 'red')
plt.bar([0.75,1.75,2.75,3.75],B_revenue,label = "COM_B",width =0.4,color = 'yellow')
plt.legend()
```
