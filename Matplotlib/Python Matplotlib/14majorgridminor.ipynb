```python
'''
@Author: Rashmi
@Date: 2021-10-7 1:40
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 2:20
@Title :Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

Date,Close
03-10-16,772.559998
04-10-16,776.429993
05-10-16,776.469971
06-10-16,776.859985
07-10-16,775.080017
'''
import matplotlib.pyplot as plt 

date=["3-10-16","4-10-16","5-10-16","6-10-16","7-10-16"]
close=[772.559998,776.429993,776.469971,776.8599858,775.080017]

plt.plot(date,close,color = 'blue',marker = '*') 
plt.xlabel("Dates")
plt.ylabel("Closing value")
plt.title("line chart of closing value of Alphabet Inc.")
# Turn on the minor TICKS, which are required for the minor GRID
plt.minorticks_on()
# Customize the major grid
plt.grid(which='major', linestyle='-', linewidth='0.5', color='blue')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
# Turn off the display of all ticks.
plt.show()
```
