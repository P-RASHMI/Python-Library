```python
'''
@Author: Rashmi
@Date: 2021-10-7 2:23
@Last Modified by: Rashmi
@Last Modified time: 2021-10-7 2:50
@Title :Write a Python program to create multiple plots.
'''
import matplotlib.pyplot as plt
import numpy as np
   
months = ['jan','feb','march','april'] 
com_A = [15,20,30,35] 
com_B = [10,11,12,13]
com_C = [5,10,15,25]
# plt.xlabel("month")
# plt.ylabel("profits")
# plt.title("company A,B,C profits")
# notice that nrows = number of images
plt.subplot(1, 3, 2)              #rows,columns,index
plt.plot(months,com_A,'r+')   
plt.subplot(1, 3, 1)             # ncols stays as 1 
plt.plot(months,com_B,'g*')
plt.subplot(1, 3, 3)                   # each image has a unique index(position)
plt.plot(months,com_C,'c^') 
plt.show()
```
