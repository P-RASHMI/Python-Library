```python
'''
@Author: Rashmi
@Date: 2021-10-6 19:54
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 20:06
@Title : Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1
'''
import matplotlib.pyplot as plt
with open("Test.txt") as textfile:
    data = textfile.read()
data = data.split('\n')
index1 = [row.split(' ')[0] for row in data]
index2 = [row.split(' ')[1] for row in data]
plt.plot(index1,index2,'rD--',linewidth = 2)
plt.xlabel('data x-axis')                                      
plt.ylabel('data y- axis')
plt.title('Test Data')                                                               
plt.show()



```
