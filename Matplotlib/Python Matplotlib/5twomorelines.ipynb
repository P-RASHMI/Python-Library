```python
'''
@Author: Rashmi
@Date: 2021-10-6 21:30
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 21:55
@Title : Write a Python program to plot two or more lines on same plot with suitable legends of each
line.
'''
import matplotlib.pyplot as plt
months = ['jan','feb','march','april']
A_revenue = [15,12,18,20]
B_revenue = [21,25,44,31]
plt.xlabel("Months")
plt.ylabel("revenue")
plt.title("Monthly revenues of Company A and B")
plt.plot(months,A_revenue,label = "COM_A")
plt.plot(months,B_revenue,label = "COM_B")
plt.legend()
```
