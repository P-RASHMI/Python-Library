```python
'''
@Author: Rashmi
@Date: 2021-10-6 23:05
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 23:22
@Title : Write a Python program to plot quantities which have an x and y position.
'''
import matplotlib.pyplot as plt 

x=[1,2,3,4,5,7]
y=[2,1,6,4,8,5]
# plt.plot(x,y,'b*')   #using plot
plt.scatter(x,y,color = 'blue',marker = 's') #using scatter()
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("quantities relating to x,y position in Matplotlib Plot")
plt.show()
```
