```python
'''
@Author: Rashmi
@Date: 2021-10-6 19:54
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 20:06
@Title : write a Python program to draw a line using given axis values with suitable label in the x axis
, y axis and a title
'''
import matplotlib.pyplot as plt
year = [1,2,3,4]
score = [88,77,66,55]
plt.plot(year,score,'r:',linewidth = 3) #plotting #- solid line ,:dotted
plt.xlabel('Year') #names for x axis and y axis
plt.ylabel('Score')
plt.title('percentage of marks')
plt.show() #to show the plot
```
