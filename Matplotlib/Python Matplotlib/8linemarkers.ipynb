```python
'''
@Author: Rashmi
@Date: 2021-10-6 22:05
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 22:25
@Title : Write a Python program to plot two or more lines and set the line markers.
'''
import matplotlib.pyplot as plt
months = ['jan','feb','march','april']
A_revenue = [15,12,18,20]
B_revenue = [21,25,44,31]
C_revenue = [49,50,80,34]
plt.xlabel("Months")
plt.ylabel("revenue")
plt.title("Monthly revenues of Company A,B,C")
plt.plot(months,A_revenue,label = "COM_A",linewidth = 5,color = 'yellow',linestyle = 'dotted',marker='o')
plt.plot(months,B_revenue,label = "COM_B",linewidth = 1,color = 'green',linestyle = 'solid',marker='>')
plt.plot(months,C_revenue,label = "COM_C",linewidth = 3,color = 'red',linestyle = 'dashed',marker='|')
plt.legend()
```
