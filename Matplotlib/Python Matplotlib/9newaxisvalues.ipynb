```python
'''
@Author: Rashmi
@Date: 2021-10-6 22:05
@Last Modified by: Rashmi
@Last Modified time: 2021-10-6 22:25
@Title : Write a Python program to display the current axis limits values and set new axis values.
'''
import matplotlib.pyplot as plt 

x=[1,2,3,4,5,7]
y=[2,1,6,4,8,5]
values = ['A', 'B', 'C', 'D','E','F'] 

plt.plot(x,y,color = 'cyan',marker="^")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Set X labels in Matplotlib Plot")
plt.xticks(x,values)   #to set new axis,replacing x data with values
plt.show()
```
