```python
'''
@Author: Rashmi
@Date: 2021-10-9 00:20
@Last Modified by: Rashmi
@Last Modified time: 2021-10-9 00:33
@Title :Write a program to draw a swarm plot of total bill against size for a given dataset tips.csv'''
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
store = sns.load_dataset('tips')
# print(store)
sns.swarmplot(x='size',y='total_bill',data=store,hue='sex') 
```
