```python
'''
@Author: Rashmi
@Date: 2021-10-9 00:35
@Last Modified by: Rashmi
@Last Modified time: 2021-10-9 00:40
@Title :Write a program to draw swarm plot of “total bill” against day for a 
dataset given in url tips.csv'''
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
store = sns.load_dataset('tips')
# print(store)
sns.swarmplot(x='day',y='total_bill',data=store) 
```
